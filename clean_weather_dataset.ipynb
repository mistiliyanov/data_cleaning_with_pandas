```python
weather_df = pd.read_csv('data/weather.csv')
```

```python
weather_df.element.unique()
```

```python
weather_df.columns
```

```python
weather_df_melted = weather_df.melt(id_vars=['id', 'year', 'month', 'element'],
                var_name='day',  # New column for day
                value_name='temperature')  # New column for temperature
```

```python
weather_df_melted
```

```python
weather_df_melted.dtypes
```

```python
weather_df_melted['day'] = weather_df_melted['day'].str.slice(1).astype(int)
```

```python
weather_df_melted.dtypes
```

```python
weather_df_melted_drop_na = weather_df_melted.dropna()
```

```python
weather_df_melted_drop_na
```

```python
weather_df_melted_drop_na['date'] =  pd.to_datetime(weather_df_melted_drop_na[['year','month', 'day']])
```

```python
weather_df_melted_drop_na.drop(columns=['id', 'year', 'month', 'day'], inplace=True)
```

```python
weather_df_melted_drop_na
```

```python
weather_df_melted_drop_na_final = weather_df_melted_drop_na.pivot(index='date', columns='element', values='temperature').reset_index()
```

```python
weather_df_melted_drop_na_final
```

```python
weather_df_melted_drop_na_final.columns
```

```python
weather_df_melted_drop_na_final.to_csv('data/weather_tidy.csv', index=None)
```
