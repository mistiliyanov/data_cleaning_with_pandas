```python
tb_melted = pd.melt(tb,
        id_vars = ['iso2', 'year'],
        value_vars= ['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'],
       var_name='gender_age_group',
       value_name='cases')
```

```python
tb_melted_na_dropped =  tb_melted.dropna()
```

```python
tb_melted_na_dropped['cases'] = tb_melted_na_dropped['cases'].astype(int)
```

```python
tb_melted_na_dropped['gender'] = tb_melted_na_dropped['gender_age_group'].str.slice(0,1)
```

```python
tb_melted_na_dropped['gender'].unique()

```

```python
tb_melted_na_dropped['age_group'] = tb_melted_na_dropped['gender_age_group'].str.slice(1)
```

```python
tb_melted_na_dropped['age_group'].unique()
```

```python
def process_age_group(age_group):
    ages = {"04": "0-14", "014": "0-14", "514": "0-14", "65": "65+", "u": "unknown"}
    if age_group in ages:
        return ages[age_group]
    else:
        # Put a dash before the last two digits
        return f"{age_group[:-2]}-{age_group[-2:]}"

process_age_group('04')
```

```python
tb_melted_na_dropped['age_group'] = tb_melted_na_dropped['age_group'].apply(process_age_group)
```

```python
tb_melted_na_dropped
```

```python
tb_melted_na_dropped.drop(columns=['gender_age_group'])
```

```python
tb_melted_na_dropped = tb_melted_na_dropped[['iso2', 'year', 'gender', 'age_group', 'cases']]
```
