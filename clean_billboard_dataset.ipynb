## Date cleaning and Tidying 

```python
bilboard = pd.read_csv('data/billboard.csv')
```

```python
billboard_melted = pd.melt(
    bilboard, 
    id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
    value_vars=[
        'wk1', 'wk2', 'wk3', 'wk4', 'wk5', 'wk6', 'wk7', 'wk8', 'wk9','wk10', 'wk11',
        'wk12','wk13', 'wk14', 'wk15', 'wk16', 'wk17', 'wk18', 'wk19', 'wk20', 'wk21',
        'wk22', 'wk23', 'wk24', 'wk25', 'wk26', 'wk27', 'wk28', 'wk29', 'wk30','wk31',
        'wk32', 'wk33', 'wk34', 'wk35', 'wk36', 'wk37', 'wk38', 'wk39','wk40', 'wk41',
        'wk42', 'wk43', 'wk44', 'wk45', 'wk46', 'wk47', 'wk48','wk49', 'wk50', 'wk51',
        'wk52', 'wk53', 'wk54', 'wk55', 'wk56', 'wk57','wk58', 'wk59', 'wk60', 'wk61',
        'wk62', 'wk63', 'wk64', 'wk65', 'wk66','wk67', 'wk68', 'wk69', 'wk70', 'wk71',
        'wk72', 'wk73', 'wk74', 'wk75','wk76'
        ],
       var_name='week',
       value_name='rank'
    )
```

```python
billboard_melted
```

```python
billboard_melted.dtypes
```

```python
billboard_melted_na_dropped = billboard_melted.dropna()
```

```python
billboard_melted_na_dropped['rank'] = billboard_melted_na_dropped['rank'].astype(int)
```

```python
billboard_melted_na_dropped.dtypes
```

```python
billboard_melted_na_dropped['week'] = billboard_melted_na_dropped['week'].str[2:].astype(int)
```

```python
billboard_melted_na_dropped = billboard_melted_na_dropped.drop(columns = 'year')
```
